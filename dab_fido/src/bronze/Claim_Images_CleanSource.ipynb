```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
base_path = "/Volumes/smart_claims_dev/00_landing/claims"
source_path = f"{base_path}/images"
archive_path = f"{base_path}/archive"
metadata_path = f"{base_path}/metadata"
```

```python
archive_configs = {
    "cloudFiles.cleanSource":"MOVE",
    "cloudFiles.cleanSource.retentionDuration":"1 minute",
    "cloudFiles.cleanSource.moveDestination":archive_path
}
```

```python
df = spark.readStream.format("cloudFiles")\
            .option("cloudFiles.format", "BINARYFILE")\
            .option("cloudFiles.schemaLocation", f"{metadata_path}/checkpoint")\
            .options(**archive_configs)\
            .load(source_path)

df.writeStream\
    .option("checkpointLocation", f"{metadata_path}/checkpoint")\
    .trigger(once=True)\
    .table("smart_claims_dev.01_bronze.claims_images")
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.readStream.format("cloudFiles")\
            .option("cloudFiles.format", "parquet")\
            .option("cloudFiles.schemaLocation", 'abfss://bronze@fidodl.dfs.core.windows.net/telematics/checkpoint')\
            .option("cloudFiles.schemaEvolutionMode", "addNewColumns")\
            .load("/Volumes/smart_claims_dev/00_landing/tele_data")

df = df.drop("_rescued_data")
```

```python
df.writeStream.format("delta")\
    .outputMode('append')\
    .option("checkpointLocation", "abfss://bronze@fidodl.dfs.core.windows.net/telematics/checkpoint")\
    .trigger(once=True)\
    .toTable("smart_claims_dev.01_bronze.telematics")
```
